### **3. Colormaps (cmap)**

O parâmetro `cmap` em funções de plotagem do **Matplotlib** define o mapa de cores utilizado no gráfico.  
Um *colormap* é basicamente uma paleta que transforma valores numéricos em cores.  

Existem diferentes tipos de colormaps:  

- **Perceptualmente uniformes:** mantêm a mesma percepção de contraste em toda a escala (ex.: `viridis`, `plasma`).  
- **Sequenciais:** indicam intensidade ou progressão de valores (ex.: `Blues`, `Greens`).  
- **Divergentes:** destacam valores em torno de um ponto médio (ex.: `coolwarm`, `bwr`).  
- **Qualitativos:** usados para distinguir categorias sem relação numérica (ex.: `Set1`, `Pastel1`).  

No exemplo, usamos `cmap="Set1"`, que é um colormap **qualitativo**, adequado para diferenciar classes ou categorias distintas.

## **➡️ Regressão Linear Simples**

A **Regressão Linear Simples** é um método estatístico utilizado para modelar a relação entre duas variáveis:
- uma variável **independente** (ou preditora), denotada por $X$
- e uma variável **dependente** (ou resposta), denotada por $y$.

O objetivo é encontrar uma equação de reta que melhor represente essa relação.

Essa equação assume a forma:

$$
\Huge y = β_0 + β_1 x + ϵ
$$

onde:
- $y$ é o valor previsto (variável dependente),
- $x$ é a variável independente,
- $β_1$ é o **coeficiente angular** (inclinação da reta),
- $β_0$ é o **coeficiente linear** (intercepto da reta com o eixo $y$).

A inclinação da reta ($β_1$) indica o quanto $y$ varia a cada unidade de variação em $X$, enquanto o intercepto ($β_0$) representa o valor de $y$ quando $x = 0$.

### **1. Importação das bibliotecas**

- `numpy (np)` → Usada para criar e manipular arrays numéricos de forma eficiente.

- `sklearn.linear_model.LinearRegression` → Classe do scikit-learn que implementa o algoritmo de Regressão Linear, usado para modelar a relação entre variáveis independentes (X) e dependente (y).

Aqui criamos um **conjunto de dados artificial** para exemplificar a regressão:

- Criamos `X`, uma matriz de 100 valores aleatórios entre 0 e 2.
- Definimos `y` usando a equação $y = 4 + 3X + ruído$, onde o ruído $(e)$ `np.random.randn(100, 1)` adiciona uma variação aleatória aos dados para simular um cenário realista.

### **4. Solução manual**

Na regressão linear simples, os coeficientes são obtidos a partir da **fórmula dos mínimos quadrados**:  

$$
\beta_1 = \frac{\sum (X - \bar{X})(y - \bar{y})}{\sum (X - \bar{X})^2}
$$  

$$
\beta_0 = \bar{y} - \beta_1 \bar{X}
$$  

onde:  
- $\bar{X}$ e $\bar{y}$ são as **médias amostrais** das variáveis $X$ e $y$.  
- O numerador da equação calcula o produto das diferenças entre cada valor e sua média.  
- O denominador representa a soma dos quadrados das diferenças em relação à média de $X$.  
- A **inclinação** $(\beta_1)$ e o **intercepto** $(\beta_0)$ formam a equação da reta de regressão.  


```python
mean_x = np.mean(X)
mean_y = np.mean(y)

mean_x, mean_y
```

```python
numer = np.sum((X - mean_x) * (y - mean_y))
denom = np.sum((X - mean_x) ** 2)

b1 = numer / denom
b0 = mean_y - b1 * mean_x

b1, b0
```

```python
y_pred_manual = b0 + b1 * X
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, y, color="blue", label="Dados reais")

plt.plot(X, y_pred_manual, color='green', label='Calculo manual')

plt.title("Exemplo de Regressão Linear")
plt.legend()
plt.show()
```

### **8. Métricas**

> **Coeficiente de Determinação $(R^2)$**  

O **$R^2$** é uma métrica que avalia a qualidade do ajuste de um modelo de regressão, indicando quanto da variabilidade dos dados é explicada pelo modelo.  

- O método `.score(X, y)` da classe `LinearRegression` retorna diretamente o valor de $R^2$.  
- O valor de $R^2$ varia entre **0 e 1**:  
  - **$R^2$ próximo de 1** → o modelo explica bem a variabilidade dos dados.  
  - **$R^2$ próximo de 0** → o modelo não consegue explicar a variabilidade dos dados.  


```python
r2_score = model.score(X, y)

print(f"Coeficiente de determinação (R²): {r2_score:.4f}")
```

> **Raiz do Erro Quadrático Médio (RMSE)**  

A **Raiz do Erro Quadrático Médio (RMSE)** é uma métrica que mede o erro médio entre os valores reais e os valores previstos por um modelo.  

Ela é bastante útil porque mantém a **mesma unidade dos dados originais** e **penaliza erros grandes mais fortemente** do que o erro absoluto.  

1. **Cálculo dos erros quadráticos:**  
   $$
   (y_{\text{real}} - y_{\text{previsto}})^2
   $$
   Esse cálculo retorna os erros elevados ao quadrado para cada previsão.  

2. **Cálculo da média dos erros quadráticos (MSE - Mean Squared Error):**  
   $$
   MSE = \frac{1}{n} \sum (y_{\text{real}} - y_{\text{previsto}})^2
   $$
   No Python, a função `mean_squared_error` da biblioteca `sklearn` pode ser utilizada para calcular o MSE.  

3. **Cálculo da raiz quadrada do MSE (RMSE):**  
   $$
   RMSE = \sqrt{MSE}
   $$  


```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y, y_pred) # Calcula o Erro Quadrático Médio (MSE)
rmse = np.sqrt(mse) # Calcula a Raiz do Erro Quadrático Médio (RMSE)

print(f"Raiz do Erro Quadrático Médio (RMSE): {rmse:.4f}")
```

## ➡️ Regressão Polinomial



A **Regressão Polinomial** é uma técnica estatística utilizada quando a relação entre as variáveis **não pode ser representada adequadamente por uma reta**.  
Nesse caso, o modelo ajusta uma **curva** que descreve melhor o comportamento dos dados.  

Assim como na regressão linear, temos:  
- uma variável **independente** (ou preditora), representada por $X$,  
- e uma variável **dependente** (ou resposta), representada por $y$.  

O objetivo é encontrar uma equação que capture padrões mais complexos entre $X$ e $y$.  

A equação da regressão polinomial tem a forma:  

$$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \dots + \beta_n x^n + \epsilon
$$  

onde:  
- $y$ é o valor previsto (variável dependente),  
- $x$ é a variável independente,  
- $\beta_0, \beta_1, \dots, \beta_n$ são os coeficientes do modelo,  
- $n$ é o grau do polinômio,  
- $\epsilon$ representa o erro (ruído) dos dados.  

Esse tipo de modelo permite ajustar **curvas** em vez de retas, sendo especialmente útil quando os dados apresentam comportamento curvilíneo ou relações mais complexas.

### **4. Treinamento do modelo**

A classe `PolynomialFeatures()` pertence ao módulo `sklearn.preprocessing` e é usada para gerar **novas variáveis derivadas de potências do atributo original**.  

A equação assume a forma:  

$$
y = b_0 + b_1x + b_2x^2 + b_3x^3 + \dots + b_nx^n
$$  

onde:  
- $y$ é o valor previsto (variável dependente),  
- $x$ é a variável independente,  
- $b_0$ é o **coeficiente linear** (intercepto),  
- $b_1, b_2, \dots, b_n$ são os coeficientes associados a cada potência de $x$ até o grau $n$.  

Ao incluir potências crescentes de $x$, o modelo consegue capturar **curvaturas e padrões mais complexos** nos dados, que uma regressão linear simples ou múltipla não representaria bem.  

Por exemplo, ao aplicar `PolynomialFeatures(degree=2)` em uma variável `x`, são criadas as colunas:  
- $x^1$ (o próprio valor de `x`),  
- $x^2$ (o valor de `x` ao quadrado).  

Com isso, o modelo de regressão linear passa a ter a capacidade de aprender curvas, aproximando funções polinomiais de grau definido.

Podemos manter `X0` e `X1` separados.

> Juntando em um único dataset  

Também é possível unir os dois conjuntos em um único dataset.  

- A função `np.vstack((A, B))` é usada para empilhar matrizes verticalmente.  
No nosso exemplo, ela une os pontos das classes 0 e 1 em uma única matriz `X`.  

- A função `np.hstack((a, b))` empilha vetores horizontalmente.  
Aqui, ela junta os rótulos das classes 0 e 1 em um único vetor `y`.  


```python
X = np.vstack((X0, X1))
y = np.hstack((y0, y1))

X[:3], y[:3]
```

```python
plt.figure(figsize=(8,5))

plot = plt.scatter(X[:,0], X[:,1], c=y, cmap="bwr", alpha=0.7)

plt.title("Dataset de Classificação - 2D (gerado aleatoriamente)") # título
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(
    handles = plot.legend_elements()[0],
    labels=["Classe 0", "Classe 1"]
)
plt.show()
```

```python
# ----------------------------
# Função para plotar fronteira + novo ponto
# ----------------------------

def plot_decision_boundary(model, X, y):
    # -----------------------------------------
    # Definir limites do gráfico
    # -----------------------------------------
    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1

    # -----------------------------------------
    # Criar grade de pontos (meshgrid)
    # - np.linspace → cria valores igualmente espaçados
    # - np.meshgrid → gera matriz de coordenadas (todas combinações x,y)
    # -----------------------------------------
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 200),
        np.linspace(y_min, y_max, 200)
    )

    # -----------------------------------------
    # Fazer previsões para cada ponto da grade
    # - ravel() → transforma matriz em vetor 1D
    # - np.c_ → concatena xx e yy em colunas (par ordenado [x,y])
    # - predict → gera a classe prevista pelo modelo
    # - reshape → volta para o formato da grade
    # -----------------------------------------
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # -----------------------------------------
    # Plotar regiões de decisão + pontos originais
    # -----------------------------------------
    plt.figure(figsize=(8,5))
    plt.contourf(xx, yy, Z, cmap="bwr", alpha=0.2)  # regiões de decisão
    plt.scatter(X[:,0], X[:,1], c=y, cmap="bwr", edgecolors="k")  # pontos reais

    plt.title("Classificação com fronteira de decisão")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")


plot_decision_boundary(model, X, y)
plt.show()
```

### **5. Fazer previsões**

```python
new_point = [4, 5]   # Altere aqui para testar
pred = model.predict([new_point])[0]
pred
```

```python
plot_decision_boundary(model, X, y)

plt.scatter(new_point[0], new_point[1], c="yellow", edgecolors="k", s=150, marker="*",
            label=f"Novo ponto → Classe {int(pred)}")
plt.legend()
plt.show()
```

# **📌 Treinamento: Divisão Treino e Teste**

O processo de **dividir os dados em conjuntos de treino e teste** é fundamental em Machine Learning.  
- O **conjunto de treino** é usado para ajustar os parâmetros do modelo, ou seja, ensinar o algoritmo a encontrar padrões nos dados.  
- O **conjunto de teste** é separado e usado apenas para avaliar o desempenho do modelo em dados que ele nunca viu, simulando situações do mundo real.  

No *scikit-learn*, essa divisão é feita pela função `train_test_split` do módulo `sklearn.model_selection`.  

Ela embaralha os dados e os separa em proporções definidas (por padrão, 75% treino e 25% teste), garantindo que possamos medir se o modelo **generaliza bem** além do conjunto usado no treinamento.  

```python
from sklearn.model_selection import train_test_split
```

### **1. Parâmetros**

A função `train_test_split()` permite controlar como os dados são divididos em treino e teste por meio de alguns parâmetros importantes:

- **`test_size`** → define o tamanho do conjunto de teste. Pode ser um número absoluto de amostras ou uma fração (ex.: `0.2` significa 20%).  
- **`train_size`** → define o tamanho do conjunto de treino. Se não for especificado, é calculado automaticamente como o complemento de `test_size`.  
- **`random_state`** → controla a semente do gerador de números aleatórios, garantindo que a divisão seja **reproduzível**.  
- **`shuffle`** → se `True` (padrão), embaralha os dados antes da divisão, evitando que a ordem original influencie o modelo.  
- **`stratify`** → permite passar um vetor de classes (ex.: `y`), garantindo que a **mesma proporção de classes** seja mantida tanto no treino quanto no teste (muito importante em classificação).  


```python
X = np.random.randn(100, 2) + np.array([2, 2])
y = np.zeros(100)

X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

> Preciso do `random_state=42`?

`np.random.seed(42)`

- Define a semente global do gerador de números aleatórios do NumPy.

- Garante que operações que dependem do NumPy (ex.: `np.random.randint`, `np.random.randn`) sejam reproduzíveis.


`random_state=42` no `train_test_split`

- É o parâmetro que controla a aleatoriedade da separação entre treino e teste.

- Internamente, ele usa o gerador de números aleatórios do scikit-learn (não necessariamente o do NumPy).

Se não fixarmos o `random_state`, cada execução pode gerar divisões diferentes.

### **2. Visualização dos dados**

```python
plt.figure(figsize=(8,5))

plt.scatter(X_train[:,0], X_train[:,1], color="blue", label="Dados treino")
plt.scatter(X_test[:,0], X_test[:,1], color="red", label="Dados teste")

plt.title("Exemplo de Classificação")
plt.legend()
plt.show()
```

## **4. Escolher e treinar um modelo**

A classe `LogisticRegression()` pertence ao módulo `sklearn.linear_model` e é usado para resolver problemas de **classificação**.  

Apesar do nome, não é um algoritmo de regressão no sentido tradicional:  
- ele estima a **probabilidade** de um dado pertencer a uma classe usando a **função sigmoide (logística)**, que transforma valores contínuos em resultados entre 0 e 1.  

Durante o treinamento (`.fit(X, y)`), o modelo encontra os parâmetros que melhor separam as classes.  
Depois, com o método `.predict(X)`, é possível classificar novas amostras de acordo com a classe mais provável.  

Alguns parâmetros comuns do construtor são:  
- **`solver="lbfgs"`** → define o algoritmo de otimização usado para ajustar os coeficientes.  
- **`max_iter=200`** → número máximo de iterações permitidas no processo de otimização (aumentado em relação ao padrão para evitar problemas de convergência).  


```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)  # Treina o modelo com os dados de treino

model
```

## **5. Fazer previsões**

```python
y_pred = model.predict(X_test)

y_pred[:10]
```

## **6. Avaliar o modelo**

**Métricas de Avaliação em Classificação:**

Após treinar e testar um modelo de classificação, é importante medir seu desempenho.  
O *scikit-learn* oferece diversas funções úteis:

- **`accuracy_score(y_true, y_pred)`** → calcula a **acurácia**, que é a proporção de previsões corretas em relação ao total de amostras.  

- **`classification_report(y_true, y_pred)`** → gera um relatório completo com as métricas principais:  
  - **Precisão (precision)**: entre os classificados como positivos, quantos realmente eram.  
  - **Revocação (recall)**: entre os que eram positivos, quantos o modelo conseguiu identificar.  
  - **F1-score**: média harmônica entre precisão e recall, útil para avaliar o equilíbrio do modelo.

- **`confusion_matrix(y_true, y_pred)`** → mostra os acertos e erros de cada classe, organizados em uma matriz quadrada.  

Essas métricas ajudam a entender não apenas se o modelo acerta bastante, mas **como ele se comporta em cada classe** do problema.  

**Support**: Não é uma métrica, mas mostra a quantidade de amostras reais de cada classe presentes no conjunto de teste.




> Accuracy

```python
# accuracy_score(y_true, y_pred) → calcula a proporção de acertos
acc = accuracy_score(y_test, y_pred)
acc
```

> Report

```python
# classification_report(y_true, y_pred) → relatório completo com precisão, recall e f1-score
report = classification_report(y_test, y_pred, target_names=iris.target_names)
print(report)
```

> Confusion Matrix

```python
# confusion_matrix(y_true, y_pred) → mostra acertos/erros por classe
cm = confusion_matrix(y_test, y_pred) #, normalize="true")
cm
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, cmap="Blues")
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.title("Matriz de Confusão")
plt.show()
```

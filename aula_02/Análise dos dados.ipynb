# **📌 CRISP-DM - Cross Industry Standard Process for Data Mining**

O **CRISP-DM (Cross Industry Standard Process for Data Mining)** é a metodologia mais utilizada para guiar projetos de ciência de dados e mineração de dados.  
Ela organiza o processo em **seis etapas principais**, que formam um ciclo iterativo:  

1. **Entendimento do Negócio:** definir claramente o problema a ser resolvido, os objetivos do projeto e os critérios de sucesso.  
2. **Entendimento dos Dados:** coletar os dados disponíveis, avaliar sua qualidade e compreender suas características. Muitas vezes essa etapa é revisitada em conjunto com o entendimento do negócio.  
3. **Preparação dos Dados:** limpar, transformar e organizar os dados para que estejam prontos para a modelagem. Inclui tratamento de valores ausentes, remoção de outliers, criação de variáveis (*feature engineering*) e normalização.  
4. **Modelagem:** selecionar os algoritmos de Machine Learning mais adequados, ajustar hiperparâmetros e treinar os modelos utilizando os dados preparados.  
5. **Avaliação:** verificar se os modelos atingem os objetivos definidos no início. Além das métricas técnicas, deve-se avaliar se as soluções fazem sentido para o contexto do negócio.  
6. **Implantação:** disponibilizar o modelo para uso real, seja em forma de relatórios, dashboards ou integração com sistemas. Essa etapa garante que o trabalho gere valor prático.  

> Importante: O CRISP-DM **não é linear**: ele é cíclico e flexível. Em qualquer etapa pode ser necessário retornar às anteriores, seja para refinar os dados, ajustar os modelos ou até redefinir objetivos de negócio.

> ***O que é um resumo estatístico descritivo e como interpretá-lo?***

Um **resumo estatístico descritivo** fornece **informações essenciais** sobre a **distribuição, dispersão e tendência central** dos dados, auxiliando na compreensão inicial do conjunto analisado.

Principais indicadores:

- **count**: Número de valores **não nulos** em cada coluna — útil para detectar **ausência de dados**.
- **mean**: A **média aritmética** dos valores — representa a **tendência central** dos dados, ou seja, o valor médio da variável..
- **std**: O **desvio padrão** — indica o **grau de dispersão** dos dados em relação à média. Quanto maior, mais espalhados estão os dados.
- **min**: O **menor valor** observado — útil para identificar **limites inferiores** e possíveis **outliers**.
- **quartis**:
  - **25% (1º quartil)**: Indica que **25% dos dados** são menores ou iguais a esse valor.
  - **50% (2º quartil ou mediana)**: Valor que separa a metade inferior e superior dos dados — uma medida robusta da centralidade.
  - **75% (3º quartil)**: Indica que **75% dos dados** são menores ou iguais a esse valor.
- **max**: O **maior valor** observado — revela **limites superiores** e possíveis **outliers**.

> ***Como interpretar um resumo estatístico:***
1. **Média próxima da mediana** → Sugere uma **distribuição simétrica**.
2. **Média distante da mediana** → Pode indicar **assimetria** (distribuição enviesada).
3. **Desvio padrão alto** → Os dados estão **muito dispersos** ao redor da média.
4. **Intervalo interquartílico (IQR = Q3 - Q1)** → Mede a **dispersão central** e ajuda a identificar **valores atípicos**.
5. **Mínimo ou máximo muito distantes dos quartis** → Indício de **outliers** ou dados extremos.
6. **Contagem baixa** → Pode sugerir **ausência ou falhas de preenchimento** em parte dos dados.

#### **➡️  Histrogramas**

> ***O que é um histograma e como interpretá-lo?***

O **histograma** é uma representação gráfica que mostra a **distribuição de frequências** de um conjunto de dados numéricos. Ele permite visualizar como os valores de uma variável estão distribuídos ao longo de intervalos (ou **bins**), facilitando a identificação de **padrões, simetrias, assimetrias e dispersão**.

Um histograma é composto por:
- **Barras verticais**: Cada barra representa um intervalo de valores (bin) e sua altura indica a **frequência absoluta** (ou relativa) de dados naquele intervalo.
- **Eixo horizontal (x)**: Mostra os intervalos dos valores da variável.
- **Eixo vertical (y)**: Indica o número de ocorrências (frequência) em cada intervalo.

> ***Como interpretar um histograma:***
1. **Distribuição simétrica (ex: forma de sino)** → Sugere que os dados seguem uma distribuição normal.
2. **Distribuição assimétrica à direita (cauda longa à direita)** → Indica que há valores mais extremos acima da média.
3. **Distribuição assimétrica à esquerda (cauda longa à esquerda)** → Indica que há valores mais extremos abaixo da média.
4. **Picos ou múltiplos modos** → Podem indicar a presença de **grupos distintos** ou **multimodalidade**.
5. **Largura dos bins**: Pode influenciar a leitura — bins muito largos podem esconder padrões, e bins muito estreitos podem gerar “ruído visual”.

> Um **histograma unimodal ou simétrico** centraliza os dados na média (medida central) e possui características por meio da distribuição da média e do desvio padrão. Uma característica do histograma unimodal ou simétrico é conter a partir do centro do gráfico o maior número de dados.

#### **➡️ Boxplots**

> ***O que é um Boxplot e como interpretá-lo?***

O **boxplot**, ou **diagrama de caixa**, é uma representação gráfica que resume a distribuição de um conjunto de dados numéricos. Ele fornece informações sobre a **dispersão, tendência central e presença de outliers** de uma variável.

Um boxplot é composto pelos seguintes elementos:
- **Caixa (box)**: Representa o intervalo interquartil (**Q1 a Q3**), que contém os **50% centrais** dos dados.
- **Linha dentro da caixa**: Indica a **mediana (Q2)**, que divide os dados em duas metades.
- **"Bigodes" (whiskers)**: Estendem-se até os valores mínimo e máximo dentro de um limite esperado, geralmente 1,5 vezes o intervalo interquartil (IQR).
- **Outliers**: Valores fora do limite dos bigodes são representados como pontos isolados.

> ***Como interpretar um boxplot:***
1. **Se a caixa é estreita** → Baixa dispersão dos dados (valores concentrados).
2. **Se a caixa é larga** → Alta dispersão dos dados (valores mais espalhados).
3. **Se a mediana está próxima de um dos extremos da caixa** → Indica assimetria na distribuição.
4. **Se há muitos outliers** → Pode sugerir a presença de valores anômalos ou uma distribuição com cauda longa.

### **7. Tratamento de Outliers**

Outliers são valores que se distanciam significativamente do padrão geral dos dados. Esta etapa busca tratá-los e refletir sobre sua origem e impacto. Dependendo do contexto, outliers podem ser removidos, transformados ou mantidos, sempre com base em critérios técnicos e objetivos analíticos.

#### **➡️ Remoção de Outliers com Base no Intervalo Interquartil (IQR)**

A técnica de remoção de outliers com base no **Intervalo Interquartil (IQR)** é uma abordagem estatística robusta que identifica e exclui valores atípicos com base na dispersão dos dados. O IQR é definido como a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1), representando a faixa central onde se encontra a maioria dos dados.

Para detectar os outliers, calcula-se um limite inferior e superior:

- **Limite inferior** = Q1 − 1.5 × IQR  
- **Limite superior** = Q3 + 1.5 × IQR

Qualquer valor fora desse intervalo é considerado um outlier e pode ser removido. Essa técnica é particularmente útil por ser insensível à influência de valores extremos, mantendo a integridade da análise estatística. Quando aplicada a grupos distintos (por exemplo, categorias de uma variável), o cálculo do IQR é feito separadamente para cada grupo, garantindo uma detecção mais justa e contextualizada dos outliers.

```python
def remove_outliers_iqr(df, column, category_col=None):
    if category_col:
        filtered_df = pd.DataFrame()
        for category in df[category_col].unique():
            subset = df[df[category_col] == category]
            Q1 = subset[column].quantile(0.25)
            Q3 = subset[column].quantile(0.75)
            IQR = Q3 - Q1
            mask = (subset[column] >= Q1 - 1.5 * IQR) & (subset[column] <= Q3 + 1.5 * IQR)
            filtered_df = pd.concat([filtered_df, subset[mask]])
        return filtered_df
    else:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        mask = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)
        return df[mask]
```

# 📌 Regressão Logística

A **Regressão Logística** é um método estatístico utilizado para modelar a **probabilidade** de ocorrência de um evento binário (por exemplo, sucesso ou fracasso, sim ou não, 0 ou 1), com base em uma variável **independente** (ou preditora), denotada por $X$.

Diferente da regressão linear, que estima diretamente um valor numérico, a regressão logística estima a **probabilidade** de que a variável dependente $y$ pertença a uma das duas classes possíveis.

A equação da regressão logística assume a forma:

$$
\Huge P(y=1|x) = \frac{1}{1 + e^{-(mx + b)}}
$$

onde:
- $P(y=1|x)$ é a **probabilidade** de que o resultado seja 1, dado o valor de $x$,
- $x$ é a variável independente,
- $m$ é o **coeficiente** que representa o impacto de $x$ na probabilidade,
- $b$ é o **intercepto** (também chamado de viés ou bias),
- $e$ é a base do logaritmo natural (aproximadamente 2,718).

A função que transforma a combinação linear $(mx + b)$ em uma probabilidade é chamada de **função sigmoide** ou **logística**, e tem a característica de mapear qualquer número real para um valor entre 0 e 1:

$$
\Huge \sigma(z) = \frac{1}{1 + e^{-z}}
$$

Assim, a regressão logística não retorna diretamente uma classe, mas sim uma **probabilidade**. Para tomar uma decisão (por exemplo, classificar como 0 ou 1), é comum utilizar um **limiar de decisão** (threshold), geralmente 0.5. Se a probabilidade for maior que 0.5, classifica-se como 1; caso contrário, como 0.

Uma forma interessante de interpretar o resultado de uma **Regressão Logística** é que a sua saída é: **a probabilidade de que a classe y vai ser igual a 1 dado uma certa entrada x**.

```python
def remove_outliers_iqr(df, column, category_col=None):
    if category_col:
        filtered_df = pd.DataFrame()
        for category in df[category_col].unique():
            subset = df[df[category_col] == category]
            Q1 = subset[column].quantile(0.25)
            Q3 = subset[column].quantile(0.75)
            IQR = Q3 - Q1
            mask = (subset[column] >= Q1 - 1.5 * IQR) & (subset[column] <= Q3 + 1.5 * IQR)
            filtered_df = pd.concat([filtered_df, subset[mask]])
        return filtered_df
    else:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        mask = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)
        return df[mask]
```

## 🔎 Métricas de Avaliação

> **Acurácia**

Proporção de previsões corretas sobre o total.
$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

> **Precisão**

Entre os preditos como positivos, quantos realmente são positivos?
$$
Precision = \frac{TP}{TP + FP}
$$

> **Recall** (Revocação ou Sensibilidade)

Entre os positivos reais, quantos foram identificados corretamente?
$$
Recall = \frac{TP}{TP + FN}
$$

> **F1-Score**

Média harmônica entre Precisão e Recall.
$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
$$

Macro vs Micro (para múltiplas classes)
- **Macro**: calcula a métrica em cada classe e faz a média simples (todas as classes têm o mesmo peso).  
- **Micro**: calcula TP, FP e FN globais antes de aplicar a fórmula (classes maiores têm mais peso).

## 🔎 Validação Cruzada

A **validação cruzada** é uma técnica utilizada em *machine learning* para avaliar a capacidade de generalização de um modelo, ou seja, seu desempenho em dados não vistos durante o treinamento. O método mais comum é o **k-fold cross-validation**, em que o conjunto de dados é dividido em *k* partes (ou "folds"). O modelo é treinado *k* vezes, cada vez utilizando uma parte diferente como conjunto de teste e as restantes como treinamento. Os resultados são então agregados para fornecer uma estimativa mais robusta da performance do modelo. Essa abordagem ajuda a evitar overfitting e fornece uma avaliação mais confiável do comportamento do modelo em produção.

```python
cv_acc_log = cross_val_score(pipe_log, X, y, cv=10, scoring='accuracy')

print(f"Acurácias da Validação Cruzada: {cv_acc_log}")
print(f"Acurácia Média CV: {np.mean(cv_acc_log):.4f}")
print(f"Desvio Padrão CV: {np.std(cv_acc_log):.4f}")
```

# 📌 KNN - K-Nearest Neighbors

O **K-Nearest Neighbors (KNN)** é um algoritmo de aprendizado supervisionado que pertence à classe dos **métodos baseados em instância**. Isso significa que ele **não constrói um modelo durante o treinamento**, apenas armazena os dados. Quando precisa fazer uma predição, o KNN compara a nova amostra com os exemplos do conjunto de treino, utilizando uma medida de distância (como a Euclidiana), e seleciona os **K vizinhos mais próximos**.

Na **classificação**, a classe mais comum entre os vizinhos é atribuída à nova amostra. Na **regressão**, a média dos valores dos vizinhos é usada.

> Por depender fortemente da similaridade entre exemplos, o KNN é sensível à escala dos dados e ao valor escolhido para K.

Por ser simples e intuitivo, o KNN é uma boa escolha para problemas onde as relações entre os dados são não lineares ou difíceis de modelar, mas pode se tornar lento e custoso em termos de memória quando aplicado a grandes volumes de dados.

## **3. Influência do valor de N**

```python
def fit_knn(n_neighbors, X_train, y_train, X_test, y_test):

    pipe = Pipeline(steps=[
        ('pre', pre_knn),
        ('reg', KNeighborsClassifier(n_neighbors=n_neighbors))
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return acc, f1
```
